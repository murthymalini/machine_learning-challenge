## Loop to transform the database of all images to arrays database

```python
import os
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array

# Path for looking all jpg files
path = './Images/'

# All path files for jpg files contained in the files list
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.jpg' in file:
            if 'gif' not in file:
                if 'bar_0527.jpg' not in file:
                    files.append(os.path.join(r, file))

# Use it for testing the code. It limits the files to 10 files
# files = files[0:10]

# Convert all images in a Database
data = []
X = []
y = []
for f in files:
    print(f)
    # Import the Image
    filepath = f
    image_size = 50
    LR = 1e-3
    # img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE) 
    img = cv2.imread(filepath, 0) 
    img = cv2.resize(img, (image_size, image_size))
    X.append(img)
    # Extract the folder name from each path
    # Get the directory name from the specified path 
    dirname = os.path.dirname(filepath)
    # Replace the initial part of the folder to leave just the folder name
    dirname = dirname.replace("./Images/", "")
    y.append(dirname)
    # data.append([np.array(X),np.array(y)])

# suffle(data)
```

```python
X = np.load('X.npy', allow_pickle=False)
X = np.array(X)
y_text = np.load('y.npy', allow_pickle=False)
y_text = np.array(y_text)


print("X Shape:", X.shape)
print("y Shape:", y_text.shape)
print("y_text:", y_text[1000])
X.shape[1:]
```

```python
from sklearn.preprocessing import LabelBinarizer
# Encode the 67 categories into binary numbers
encoder = LabelBinarizer()
y = encoder.fit_transform(y_text)
print("y:", y[1000])
# Sample to decode the binary encoding
inverse = encoder.inverse_transform(y)
print(inverse[1000])
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
print(y_train[0])
print(y_train[1])
print("X_train Shape:", X_train.shape)
print("y_train Shape:", y_train.shape)
print("X_test Shape:", X_test.shape)
print("y_test Shape:", y_test.shape)
```

```python
X_train
```

```python
X_train = X_train.reshape(11684,50,50,1)
X_test = X_test.reshape(3895,50,50,1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
model = Sequential()
```

```python
# input_shape=img_data[0].shape
# print(input_shape)
```

```python
model.add(Conv2D(96, (1, 1), activation='relu', input_shape=(50,50,1)))
```

```python
model.add(MaxPooling2D(pool_size=(3, 3)))
```

```python
model.add(Conv2D(64, (1, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(3, 3)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(100, activation='relu'))
```

```python
model.add(Dense(67, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(X_train, y_train, 
           batch_size=32, epochs=100, validation_split=0.01)
```

```python
model.evaluate(X_test, y_test)[1]
```
